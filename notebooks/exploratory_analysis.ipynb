Analysis of mental health specialist density, distribution, and access gaps across Iberoamerican countries.

````python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns, warnings
warnings.filterwarnings('ignore'); sns.set_theme(style='whitegrid'); print('Libraries loaded')
````

````python
import os
df=None
for p in ['../data/mental_health_specialists.csv','../data/data.csv']:
    if os.path.exists(p): df=pd.read_csv(p); break
if df is None:
    df=pd.DataFrame({'country':['Spain','Mexico','Argentina','Brazil','Colombia','Chile','Peru','Portugal','Ecuador','Uruguay'],'psychiatrists_per_100k':[11.2,1.6,5.8,3.2,1.8,4.1,0.9,10.5,0.7,4.3],'psychologists_per_100k':[20.5,3.1,114.0,5.3,4.2,8.2,2.1,7.5,1.5,27.8],'region':['Europe','LAC','LAC','LAC','LAC','LAC','LAC','Europe','LAC','LAC']})
    print('Using sample data')
print(df.shape); df.head()
````

````python
fig,axes=plt.subplots(1,2,figsize=(13,5))
df.sort_values('psychiatrists_per_100k',ascending=True).plot.barh(x='country',y='psychiatrists_per_100k',ax=axes[0],legend=False,color='steelblue')
axes[0].set_title('Psychiatrists per 100k Population'); axes[0].axvline(df['psychiatrists_per_100k'].mean(),color='red',linestyle='--',label=f'Mean: {df["psychiatrists_per_100k"].mean():.1f}')
axes[0].legend()
df.sort_values('psychologists_per_100k',ascending=True).plot.barh(x='country',y='psychologists_per_100k',ax=axes[1],legend=False,color='coral')
axes[1].set_title('Psychologists per 100k Population')
plt.tight_layout(); plt.show()
````
