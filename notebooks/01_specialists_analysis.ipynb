````python
import latamdata as ld
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (14, 7)
plt.rcParams['figure.dpi'] = 120
sns.set_theme(style='whitegrid')
print(f'latamdata-py v{ld.__version__}')
````

## 1. Load the Dataset

````python
df = ld.load('mental-health-specialists-iberoamerica')
print(f'Shape: {df.shape}')
print(f'Countries: {len(df["country"].unique())}')
print(f'Year range: {df["year"].min()} – {df["year"].max()}')
df.describe()
````

## 2. Country Ranking — Psychiatrists per 100k (2024)

````python
df_2024 = df[df['year'] == 2024].sort_values('psychiatrists_per_100k', ascending=True)

fig, ax = plt.subplots(figsize=(10, 10))
colors = ['#d62728' if v < 1.0 else '#2ca02c' if v >= 3.0 else '#1f77b4'
          for v in df_2024['psychiatrists_per_100k']]
bars = ax.barh(df_2024['country_name'], df_2024['psychiatrists_per_100k'], color=colors)

# WHO minimum threshold
ax.axvline(x=1.0, color='red', linestyle='--', alpha=0.7, label='WHO minimum (1 per 100k)')
ax.axvline(x=3.0, color='green', linestyle='--', alpha=0.7, label='WHO recommended (3 per 100k)')

ax.set_title('Psychiatrists per 100,000 population — Ibero-America (2024)', fontsize=13, fontweight='bold')
ax.set_xlabel('Psychiatrists per 100,000 population')
ax.legend()
plt.tight_layout()
plt.savefig('psychiatrists_ranking_2024.png', dpi=150, bbox_inches='tight')
plt.show()
print(f'\nCountries below WHO minimum (1/100k): {(df_2024["psychiatrists_per_100k"] < 1.0).sum()}')
````

## 3. Regional Trends Over Time

````python
# Top 5 and bottom 5 countries by 2024 rate
top5 = df_2024.nlargest(5, 'psychiatrists_per_100k')['country'].tolist()
bottom5 = df_2024.nsmallest(5, 'psychiatrists_per_100k')['country'].tolist()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

for country in top5:
    cdf = df[df['country'] == country].sort_values('year')
    ax1.plot(cdf['year'], cdf['psychiatrists_per_100k'], marker='o', markersize=4,
             label=cdf['country_name'].iloc[0])
ax1.set_title('Top 5 Countries — Psychiatrists per 100k', fontweight='bold')
ax1.set_xlabel('Year'); ax1.set_ylabel('Per 100,000')
ax1.legend(); ax1.axhline(y=3.0, color='green', linestyle='--', alpha=0.5)

for country in bottom5:
    cdf = df[df['country'] == country].sort_values('year')
    ax2.plot(cdf['year'], cdf['psychiatrists_per_100k'], marker='o', markersize=4,
             label=cdf['country_name'].iloc[0])
ax2.set_title('Bottom 5 Countries — Psychiatrists per 100k', fontweight='bold')
ax2.set_xlabel('Year'); ax2.set_ylabel('Per 100,000')
ax2.legend(); ax2.axhline(y=1.0, color='red', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig('specialist_trends.png', dpi=150, bbox_inches='tight')
plt.show()
````
